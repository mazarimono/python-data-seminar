## ZIPファイルを操作する

```python
# パスをコピペして貼り付け、プリント
a_path = '/content/drive/MyDrive/od-book/国勢調査/A002005212020DDSWC27.zip'
t_path = '/content/drive/MyDrive/od-book/国勢調査/tblT001082C27.zip'
print(a_path)
print(t_path)
```

```python
# Pathオブジェクトにして、変数に渡す
toukei_zip_path = Path(t_path)
kyokai_zip_path = Path(a_path)

# 様々な操作が容易になる
print(toukei_zip_path.name)
print(toukei_zip_path.stem)
print(toukei_zip_path.parent)
```

### コード5-1

```python
with ZipFile(kyokai_zip_path, 'r') as kyokai_zip:
    kyokai_zip_name = kyokai_zip.infolist()
print(kyokai_zip_name)
```
