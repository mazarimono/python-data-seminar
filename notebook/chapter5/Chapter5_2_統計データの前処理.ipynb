```python
print(type(df.loc[0, '総数、年齢「不詳」含む']))
print('--- --- ---')
print(df['総数、年齢「不詳」含む'].values) #①
print(' --- --- ---')
print(df['総数、年齢「不詳」含む'].value_counts().head()) #②
```

```python
# 人数が入るカラムに含まれる数値以外の文字データを探る
except_list = list()  #① 数値以外の文字列を格納するリスト
for col in df.loc[:, '総数、年齢「不詳」含む':].columns:
    s_v = df[col].values #② 選択された列のすべての値
    for w in s_v: #③ 上のすべての値をループ
        try: #④ 文字列を数値にできるか？
            int(w)
        except ValueError: # ⑤出来なければ、except_listに格納
            except_list.append(w)
cnt_exceptions = Counter(except_list) #⑥ Counterを使って要素を数える
print(f'文字データのカウント結果: {cnt_exceptions}') #⑦ f-stringで数値型に出力できなかった文字列のカウントを出力
```

## 人数を数値に変換する

```python
for col in df.loc[:, '総数、年齢「不詳」含む':].columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
```

## 処理をまとめた関数を作成する

```python
def prepro_kokusei(data_path: str) -> pd.DataFrame:
    '''
    国勢調査の統計データを前処理する関数
    Params:
        data_path: str
        zip_fileのあるパス
    Returns:
        df: pd.DataFrame
        前処理済みのデータ

    '''
    df = pd.read_csv(data_path, encoding='cp932', header=[0, 1])
    front_col = [col for col in df.columns.get_level_values(0) if not col.startswith('T00')]
    back_col = [col for col in df.columns.get_level_values(1) if not col.startswith("Unnamed:")]
    new_col = front_col + back_col
    df.columns = new_col
    for col in df.loc[:, '総数、年齢「不詳」含む':].columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    return df
```

## 書籍にはありませんが、カラム名を指定して国勢調査の異なるデータも前処理できる関数を作る

- そして異なるデータで動作を試す

```python
def prepro_kokusei(data_path: str, col_name) -> pd.DataFrame:
    '''
    国勢調査の統計データを前処理する関数
    Params:
        data_path: str
        zip_fileのあるパス
    Returns:
        df: pd.DataFrame
        前処理済みのデータ

    '''
    df = pd.read_csv(data_path, encoding='cp932', header=[0, 1])
    front_col = [col for col in df.columns.get_level_values(0) if not col.startswith('T00')]
    back_col = [col for col in df.columns.get_level_values(1) if not col.startswith("Unnamed:")]
    new_col = front_col + back_col
    df.columns = new_col
    for col in df.loc[:, col_name:].columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)
    return df
```
