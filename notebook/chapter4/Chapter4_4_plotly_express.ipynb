```python
import plotly.express as px
import plotly.data as data
```

```python
gapminder = data.gapminder() # データを呼び出し、変数に格納
fig = px.line(gapminder, x='year', y='gdpPercap', color='country', title='chart') # line関数でグラフを作成。変数に格納
fig.show() # showでグラフを表示。
```

```python
fig = px.box(gapminder, x='year', y='gdpPercap', points='all', hover_name='country')
fig.show()
```

```python
fig = px.box(gapminder, x='year', y='gdpPercap', points='all', hover_name='country', color='continent')
fig.show()
```
