```python
from shapely.geometry import Point, LineString, Polygon, MultiPoint, MultiLineString, MultiPolygon, GeometryCollection
import folium
```

```python
point1 = Point(135.758, 34.985) # 経度、緯度の順で渡す
point2 = Point(135.729, 35.039)
point3 = Point(135.746, 35.066)
point4 = Point(135.798, 35.027)
multip = MultiPoint([point1, point2, point3, point4]) #複数のポイントを格納
print(point1)
print(multip)
```

```python
center = multip.centroid #①
m = folium.Map([center.y, center.x], zoom_start=12) #②
folium.Marker([center.y, center.x], icon=folium.Icon(color='red')).add_to(m) #③
folium.GeoJson(multip).add_to(m) #④
m #⑤
```

```python
line1 = LineString([point1, point3]) #①
line2 = LineString([point2, point4])
line3 = LineString([point1, point2, point3])
multil = MultiLineString([line1, line2])


intersection_point = line1.intersection(line2) #②
center = multil.centroid
m = folium.Map([center.y, center.x], zoom_start=12)
folium.Marker([intersection_point.y, intersection_point.x]).add_to(m) #③
folium.GeoJson(multil).add_to(m)
folium.GeoJson(line3, style_function=lambda feature:{'color': 'red'}).add_to(m)
m
```

```python
poly1 = Polygon([point1, point2, point3, point4])
center = poly1.centroid
poly_rect = poly1.minimum_rotated_rectangle # ①

m = folium.Map([center.y, center.x], zoom_start=12)
folium.GeoJson(poly1).add_to(m)
folium.GeoJson(poly_rect, style_function=lambda feature: {'color': 'green'}).add_to(m)
m
```

```python
poly_bounds = poly1.bounds
poly_rect = poly1.minimum_rotated_rectangle

print(poly_bounds)

center = poly1.centroid
m = folium.Map([center.y, center.x], zoom_start=11)
folium.GeoJson(poly1).add_to(m)
folium.GeoJson(poly_rect, style_function=lambda feature: {'color': 'green'}).add_to(m)

m
```

```python
geo_col = GeometryCollection([poly1, multil, multip])

center = poly1.centroid
m = folium.Map([center.y, center.x], zoom_start=13)
folium.GeoJson(geo_col).add_to(m)
m
```
